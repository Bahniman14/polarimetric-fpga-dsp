```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
N = 100000
x_uniform = np.random.uniform(low=0, high=1, size=N)
print(x_uniform)
plt.figure(figsize=(10, 6))


# Plot 'x' values on the y-axis, and their index on the x-axis
# s=0.5 makes the dots small, alpha=0.1 makes them transparent so you can see density
plt.scatter(range(N), x_uniform, s=0.5, alpha=1, label='Random Uniform Data')

# 3. Add labels and title
plt.title(f"Scatter Plot of {N} Samples from Uniform Distribution")
plt.xlabel("Index (0 to N)")
plt.ylabel("Value of x")
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Show the plot
plt.show()


plt.figure(figsize=(8, 4))
plt.hist(x_uniform, bins=50, color='skyblue')
plt.title("Histogram: Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
N = 100000
x_normal = np.random.normal(loc=0.0, scale=1.0, size=N)
print(x_normal)
plt.figure(figsize=(10, 6))

# Plot 'x' values on the y-axis, and their index on the x-axis
# s=0.5 makes the dots small, alpha=0.1 makes them transparent so you can see density
plt.scatter(range(N), x_normal, s=0.5, alpha=1, label='Random Normal Data')

# 3. Add labels and title
plt.title(f"Scatter Plot of {N} Samples from Normal Distribution")
plt.xlabel("Index (0 to N)")
plt.ylabel("Value of x")
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Show the plot
plt.show()


plt.figure(figsize=(8, 4))
plt.hist(x_normal, bins=100, color='skyblue', density = True)
plt.title("Histogram: Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
N = 10000000
x_poisson = np.random.poisson(lam=5, size=N)
print(x_poisson)
plt.figure(figsize=(10, 6))

# Plot 'x' values on the y-axis, and their index on the x-axis
# s=0.5 makes the dots small, alpha=0.1 makes them transparent so you can see density
plt.scatter(range(N), x_poisson, s=0.5, alpha=1, label='Random poisson Data')

# 3. Add labels and title
plt.title(f"Scatter Plot of {N} Samples from poisson Distribution")
plt.xlabel("Index (0 to N)")
plt.ylabel("Value of x")
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Show the plot
plt.show()


plt.figure(figsize=(8, 4))
plt.hist(x_poisson, bins=100, color='skyblue')
plt.title("Histogram: poisson Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
N = 1000
x_standard_cauchy = np.random.standard_cauchy(size=N)
# /print(x_standard_cauchy)
plt.figure(figsize=(10, 6))

# Plot 'x' values on the y-axis, and their index on the x-axis
# s=0.5 makes the dots small, alpha=0.1 makes them transparent so you can see density
plt.scatter(range(N), x_standard_cauchy, s=0.5, alpha=1, label='Random Lorentzian Data')

# 3. Add labels and title
plt.title(f"Scatter Plot of {N} Samples from Lorentzian Distribution")
plt.xlabel("Index (0 to N)")
plt.ylabel("Value of x")
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Show the plot
plt.show()


plt.figure(figsize=(8, 4))
plt.hist(x_standard_cauchy, bins=50, color='skyblue', density = True)
plt.title("Histogram: Lorentzian Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
N = 100000
t = np.linspace(0, 2 * np.pi, N)
y_sin = np.sin(t)

# --- Plot 1: The Sine Wave ---
plt.figure(figsize=(10, 4))
plt.plot(t, y_sin, color='blue', linewidth=2)
plt.title("Sine Wave: Value vs. Time")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Plot 2: Histogram with the "Following" Line ---
plt.figure(figsize=(10, 6))

# Plot histogram (density=True ensures the line and bars use the same scale)
plt.hist(y_sin, bins=100, density=True, color='skyblue', alpha=0.5, edgecolor='white', label='Histogram')

plt.title("Histogram of Sine Wave: Notice the U-Shape")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
